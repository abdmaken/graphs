## set a title

```python
import seaborn as sns
import matplotlib.pyplot as plt

# import dataset
phool= sns.load_dataset("iris")
phool
# draw a Line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Flowers' Plot")
plt.show()
```

## set limits

```python
import seaborn as sns
import matplotlib.pyplot as plt

# import dataset
phool= sns.load_dataset("iris")
phool
# draw a Line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Flowers' Plot")
plt.xlim(0)
plt.ylim(0)
plt.show()
```

## change figure size

```python
import seaborn as sns
import matplotlib.pyplot as plt

# import dataset
phool= sns.load_dataset("iris")
phool

# draw a Line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Flowers' Plot")
sns.set(style="ticks")
# change figure size
plt.figure(figsize=(10, 6))
plt.show()
```
